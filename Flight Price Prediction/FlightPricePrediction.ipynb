## Preproccessing the Dataset

```python
df = df.drop("Unnamed: 0", axis=1)
df = df.drop("flight", axis=1)

df["class"] = df["class"].apply(lambda x: 1 if x == "Business" else 0)
df.stops = pd.factorize(df.stops)[0]
```

```python
df = df.join(pd.get_dummies(df.airline, prefix="airline")).drop("airline", axis = 1)
df = df.join(pd.get_dummies(df.source_city, prefix="departure")).drop("source_city", axis = 1)
df = df.join(pd.get_dummies(df.destination_city, prefix="destination")).drop("destination_city", axis = 1)
df = df.join(pd.get_dummies(df.departure_time, prefix="departure")).drop("departure_time", axis = 1)
df = df.join(pd.get_dummies(df.arrival_time, prefix="arrival")).drop("arrival_time", axis = 1)
```

```python
df
```

## Training a Regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

x, y = df.drop("price", axis = 1), df.price
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.2)
```

```python
reg = RandomForestRegressor(n_jobs = -1)
reg.fit(xTrain, yTrain)
```

## Evaluating the Regression Model

```python
reg.score(xTest, yTest)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import math
import matplotlib.pyplot as plt

yPred = reg.predict(xTest)

print("R2 Score:", r2_score(yTest, yPred))
print("Mean Absolute Error:", mean_absolute_error(yTest, yPred))
print("Mean Squared Error:", mean_squared_error(yTest, yPred))
print("Root Mean Squared Error:", math.sqrt(mean_squared_error(yTest, yPred)))
```

```python
plt.scatter(yTest, yPred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Flight Price")
plt.title("Prediction vs Actual Price")
```
